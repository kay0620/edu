```python
import folium

#location : 위치
#zoom_start : 초기 화면의크기
#tiles : 지도의 유형 (기본값 : OpenStreetMap)
#        Stamen Toner,Stamen Watercolor,Stamen Terrain, CartoDB Positron
m = folium.Map(location=[35.19600417088537, 129.0815799388583], 
           zoom_start = 17,
           tiles = 'CartoDB Positron',
           width = 750, height = 500
          )

m.save('busan.html')
```

## 마커생성

folium.Marker

```python
m = folium.Map(location = [35.19600417088537, 129.0815799388583],zoom_start=15)

folium.Marker([35.19600417088537, 129.0815799388583],
              icon = folium.Icon(icon = 'thumbs-up', color ='red'),
              tooltip = 'Click me',
              popup = folium.Popup('KPC<br> 서부산', min_width = 50, max_width = 100)
             ).add_to(m)

m
```

## 원형마커

folium.CircleMarker

```python
m = folium.Map(location = [35.19600417088537, 129.0815799388583],zoom_start=15)

popuptext = folium.Popup('KPC<br> 서부산', min_width = 50, max_width = 100)

folium.CircleMarker([35.19600417088537, 129.0815799388583], radius=50, color = 'red',
                    fill = True, fill_color = '#3186cc',
                    popup = popuptext, tooltip = 'Click me'
                   ).add_to(m)

m
```

## 마커 클러스터

여러개의 마커를 그룹화 
<br>
MarkerCluster()

```python
from folium.plugins import MarkerCluster

#지도 생성
m = folium.Map(location = [35.19600417088537, 129.0815799388583],zoom_start=17)

#마커클러스터 생성
marker_cluster = MarkerCluster()

#FontAwesome 아이콘 폰트 라이브러리
#fontawesome.com
icon = folium.Icon(prefix = 'fa', icon = 'poo', color = 'green')

#마커 생성
markers = [
    folium.Marker(location=[35.19643236288122, 129.08104744712014], icon = icon, tooltip='스타벅스'),
    folium.Marker(location=[35.195766509397195, 129.08098929038914]),
    folium.Marker(location=[35.19713142346524, 129.0834969704145]),
    folium.Marker(location=[35.195602471432714, 129.085154023814]),
    folium.Marker(location=[35.19512189726322, 129.0861105459426])
]

#마커클러스터에 마커를 추가
for marker in markers:
    marker_cluster.add_child(marker)

#지도에 마커클러스터를 추가
m.add_child(marker_cluster)    

m
```
