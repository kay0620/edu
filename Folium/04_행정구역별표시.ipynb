folium.GeoJson

```python
import json
busan_geo = json.load(open('busan_gu.json', encoding = 'utf-8'))
busan_geo
```

```python
import requests
import json

r = requests.get('https://raw.githubusercontent.com/kay0620/edu/main/folium/busan_gu.json')
c = r.content
busan_geo = json.loads(c)
busan_geo
```

```python
import folium
m = folium.Map(location = [35.179774, 129.075004], zoom_start = 10, tiles = 'cartodbpositron')

folium.GeoJson(busan_geo, name = '지역구').add_to(m)
m
```

## Choropleth

지리적 경계에 따라 데이터를 시각화 할때 사용

```python
import folium

m = folium.Map(location = [35.19600417088537, 129.0815799388583],zoom_start=12,tiles = 'cartodbpositron')


data = {
    '해운대구': 9904312,
    '강서구': 3448737,
    '연제구': 2466052,
    '북구': 2890451,
    '사하구': 1529344,
    '남구': 3529344,
    '금정구': 5534234,
    '기장군': 529344,
    '동래구': 4466052
    }

m.choropleth(geo_data = busan_geo,
             data = data,
             fill_color = 'YlOrRd',
             line_opacity = 0.2,
             key_on = 'properties.name',
             legend_name = '부산시 관광객수')

m
```
