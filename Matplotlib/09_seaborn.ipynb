```python
import pandas as pd
df = pd.read_csv('../Pandas/score.csv', encoding = 'utf-8-sig')
df
```

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=df, x='이름', y='키')

plt.title('학생별 키 데이터')
plt.ylim(160,210)
plt.show()
```

## 가로 막대 그래프

```python
sns.barplot(data=df, y='이름', x='키')
plt.xlim(160,210)
plt.show()
```

## countplot

데이터의 개수를 세어 막대그래프로 표시

```python
df['학교'].unique()
df['학교'].value_counts()
```

```python
sns.countplot(data=df, x='학교')
plt.show()
```

```python
data = {
    'Year': [2010,2011,2012,2013,2014,2015],
    'Winner' : [10,15,12,18,20,14]
}

df = pd.DataFrame(data)
```
