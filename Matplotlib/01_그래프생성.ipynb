### 한글 폰트 설정

```python
#font 확인

import matplotlib.font_manager as fm

for f in fm.fontManager.ttflist:
    if f.name.find('HY') != -1:
        print(f.name)
```

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [2,4,8]
plt.plot(x,y)
plt.title('꺾은선 그래프')
plt.show()
```

```python
x = [-1,0,1]
y = [-5,-1,2]

plt.plot(x,y)
plt.title('꺾은선 그래프', size = 30, fontname = 'HYMyeongJo-Extra', fontweight = 'bold') #개별 폰트 설정

#x,y축 레이블 표시
plt.xlabel('x축', color = 'red', loc = 'right') #left, center, right
plt.ylabel('y축', color = '#22A699', loc = 'top') #top, center, bottom

# x,y 눈금
plt.xticks([-1,0,1])
plt.yticks([-5, -3, -1, 1, 3])

#범례
plt.legend()

plt.show()
```

```python
x = [1,2,3,4]
y = [-20,-10,10,20]

plt.plot(x,y, label = '온도데이터')
# plt.legend(loc = 'lower right')
#upper lower
#left center right
plt.legend(loc = (0.6,0.3)) #0~1
plt.show()
```
