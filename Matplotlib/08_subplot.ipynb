# Subplot

여러개의 그래프를 하나의 그림에 배열 형태로 그려줌

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

## 2개 그래프

```python
import matplotlib.pyplot as plt

c = [1,2,3,4,5]
math_scores = [80,70,90,85,75]

month = ['Jan','Feb','Mar','Api', 'May']
math_means = [80.5, 75.2, 50.2, 60.2, 70]

fig, axes = plt.subplots(2,1, figsize=(6,8))

#첫번째 그래프
axes[0].bar(c, math_scores)
axes[0].set_title('학급별 수학 점수')
axes[0].set_xlabel('Class')
axes[0].set_ylabel('수학 점수')


#두번째 그래프
axes[1].plot(month, math_means)
axes[1].set_title('월별 수학 점수')
axes[1].set_xlabel('month')
axes[1].set_ylabel('수학 점수')

#그래프간 간격 조정
plt.tight_layout()

plt.show()
```

## 4개 그래프

```python
import pandas as pd
df = pd.read_csv('../Pandas/score.csv', encoding = 'utf-8-sig')
df
```

```python
fig, axs = plt.subplots(2,2, figsize=(15,10))

#첫번째 그래프 
axs[0,0].bar(df['이름'],df['국어'], color = 'r')
axs[0,0].set_facecolor('lightyellow')
axs[0,0].grid(linestyle = ':')
axs[0,0].set_title('학생별 국어점수')

#두번째 그래프 
axs[0,1].plot(df['이름'],df['수학'], label = '수학')
axs[0,1].plot(df['이름'],df['영어'], label = '영어')
axs[0,1].legend()
axs[0,1].set_title('학생별 수학,영어 점수')

#세번째 그래프 
axs[1,0].barh(df['이름'],df['과학'])
axs[1,0].set_title('학생별 과학점수')

#네번째 그래프 
axs[1,1].plot(df['이름'],df['사회'], color = 'g', marker='o')
axs[1,1].set_title('학생별 사회점수')

plt.tight_layout()
plt.show()
```
