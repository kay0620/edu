## 세로 막대 그래프

```python
import matplotlib.pyplot as plt

names = ['유재석', '정형돈', '박명수']
height = [180,177,175]

# plt.bar(names, height, color = 'g')
plt.bar(names, height, color = ['r','g','b'], alpha = 0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

names = ['유재석', '정형돈', '박명수']
height = [180,177,175]

bar = plt.bar(names, height, color = ['r','g','b'], alpha = 0.7, hatch = '/', width = 0.5)

plt.ylim(170,185)

# for i, text in enumerate(height):
#     plt.text(i, text, text, ha = 'center', va = 'bottom')

#3.4이상
plt.bar_label(bar, label_type = 'edge') #center, edge

plt.show()
```

```python
bar = plt.bar(names, height)
bar[0].set_alpha(0.7)
bar[0].set_color('green')
bar[0].set_hatch('x')
```

```python
print(plt.matplotlib.__version__)
```

## 가로 막대 그래프

```python
barh = plt.barh(names, height, height = 0.5)
plt.xlim(170,185)

#text 출력
# for i, text in enumerate(height):
#     plt.text(text, i, text, va = 'center')
    
plt.bar_label(barh, labels = height)   

plt.show()
```

## 다중 막대 그래프 

Pandas.DataFrame.plot()

```python
import pandas as pd

citycar = pd.read_csv('citycar.csv', encoding='utf-8-sig', index_col = 'type')
citycar
```

```python
citycar.plot(kind='bar')
plt.show()
```

```python
citycar = citycar.T
```

```python
citycar
```

```python
citycar.plot(kind='bar', rot=0, title = '지역별 차량 등록 대수')
plt.legend(ncol=3, fontsize=10)
plt.ylim(0,90)
plt.show()
```

## 누적막대 그래프

```python
citycar.plot(kind='bar',stacked=True)
plt.show()
```
