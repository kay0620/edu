# 원그래프

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.pyplot as plt

# 프로그래밍 언어 선호도
values = [30,25,20,13,10,2]
labels= ['Python','Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels = labels,  autopct='%.1f%%',counterclock=False, startangle=90)
plt.title('언어별 선호도')

#범례
plt.legend(loc=(1.2,0.5), title='프로그래밍언어') 

plt.show()
```

```python
import matplotlib.pyplot as plt

# 프로그래밍 언어 선호도
values = [30,25,20,13,10,2]
labels= ['Python','Java', 'Javascript', 'C#', 'C/C++', 'ETC']
explode = [0.1,0,0,0,0,0]
color = ['#22A699','#F2BE22','#F29727','#F24C3D','#17594A','#213363']

plt.pie(values, labels = labels,  autopct='%.1f%%',counterclock=False, startangle=90,
        explode = explode, colors = color)
plt.title('언어별 선호도')

#범례
plt.legend(loc=(1.2,0.5), title='프로그래밍언어') 

plt.show()
```

## 도넛차트

```python
plt.pie(values, labels = labels, autopct = '%.1f%%', 
        wedgeprops = {'width': 0.6}, explode = [0.05] *6)
plt.show()
```

```python
def custom_autopct(pct):
    if pct >= 10 :
        return '%.2f%%'%(pct)
    else:
        return ''

plt.pie(values, labels = labels, autopct = custom_autopct, pctdistance=0.7,
        wedgeprops = {'width': 0.6, 'linewidth' : 2, 'edgecolor':'w' })
plt.show()
```

## 데이터프레임

score.csv 파일을 읽어와 북산고와 능남고 학생들 수를 파이차트로 표현

```python
import pandas as pd

df = pd.read_csv('../Pandas/score.csv', encoding='utf-8-sig', index_col = '지원번호')
df
```

```python
st = df.groupby('학교')['이름'].count()

print(st.index, st.values)
```

```python
plt.pie(st.values, labels = st.index,  autopct='%.1f%%',counterclock=False, startangle=90)
plt.show()
```

```python
st.plot(kind='pie', labels=st.index, autopct = '%.1f%%')
plt.legend(loc=(1,0.5), title='학교')
plt.axis('off')
plt.title('학교별 학생수')
plt.show()
```
