# 그래프 스타일 설정

- 선스타일
- 마커모양
- 그리드
- 텍스트 표시


```python
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [-20,-10,10,20]
plt.plot(x,y)
plt.show()
```

## 선스타일

```python
#선을 두껍게
# plt.plot(x,y, linewidth = 5)

#선 색깔
# plt.plot(x,y, color = 'pink')
# plt.plot(x,y,color = '#F24C3D')

# plt.plot(x,y,color = 'g',linewidth = 5) #green 

#선스타일
plt.plot(x,y, linestyle= ':',color = 'g',linewidth = 5)
# '-' 실선
# ':' 작은 점선
#'--' 큰점선
#'-.' 큰점선 + 작은점선

plt.show()
```

## 마커 모양

```python
# plt.plot(x,y, marker='v', linestyle = 'None', markersize=10)

#마커 색깔
plt.plot(x,y, marker='o', markersize=20,
        markerfacecolor = 'yellow',
        markeredgecolor = 'red',
        alpha = 0.2) #투명도 (0~1)

plt.plot([1,2,3,4],[-18,-8,5,15], marker='v')
plt.show()
```

## 스타일 포맷 설정

```python
# plt.plot(x,y, 'v:r') #선, 마커, 색깔

# plt.plot(x,y, marker='v', linestyle = ':', color = 'r')
plt.plot(x,y, marker='v', ls = ':', c = 'r', mec = 'y', ms = 10)

# fmt = '[marker][line][color]'
plt.show()
```

## 그리드

```python
plt.plot(x,y)

# 방향 설정
# x,y, both
# plt.grid(axis = 'y') 

#색깔
# plt.grid(color='#F24C3D')
# plt.grid(color='yellow')
# plt.grid(color='g', alpha = 0.2) #투명도 설정

#라인스타일
# plt.grid(linestyle=':', linewidth = 2)

plt.grid(axis='y', color='g', alpha = 0.5, linestyle = ':')
plt.grid(axis='y', c='g', alpha = 0.5, ls = ':')

plt.show()
```

## 텍스트

```python
x,y
```

```python
plt.plot(x,y, marker = 'o')
plt.grid(axis='y', c='g', alpha = 0.5, ls = ':')
plt.text(x[0],y[0], y[0])
plt.text(x[1],y[1], y[1])
plt.text(x[2],y[2], y[2])
plt.text(x[3],y[3], y[3])
plt.show()
```

```python
plt.plot(x,y, marker='o')

for i in range(4):
    plt.text(x[i],y[i], y[i], ha = 'center', va = 'bottom')
    #horizontalalignment or ha : {'left', 'center', 'right'}
    #verticalalignment or va : {'bottom', 'baseline', 'center', 'center_baseline', 'top'}
    
plt.yticks([-20,-10,0,10,20,30])
plt.show()
```

```python
plt.plot(x,y, marker='o')

for i in range(4):
    plt.text(x[i]-0.1,y[i] + 2, y[i])
    
plt.yticks([-20,-10,0,10,20,30])
plt.show()
```
