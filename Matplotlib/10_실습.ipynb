```python
data = {
    '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}

import pandas as pd

df = pd.DataFrame(data)
df
```

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

## 1. 영화 데이터를 활용하여 막대 그래프를 출력하세요

- x축 : 영화
- y축 : 평점
- x,y축 레이블 표시
- 제목 : 국내 Top8 영화 평점 정보
- x축 영화제목 : 90도 회전

```python
import matplotlib.pyplot as plt
plt.bar(df['영화'], df['평점'])
plt.xlabel('영화')
plt.ylabel('평점')
plt.title('국내 Top8 영화 평점 정보')
plt.xticks(rotation=90)
plt.show()
```

## 2. 개봉 연도별 평점 변화 추이를 꺾은선 그래프로 그리세요

- x축 : 개봉연도
- y축 : 평점
- 마커 : o
- x축 눈금 : 5년 단위 (2005,2010,2015,2020)
- y축 범위 : 최소 7, 최대 10

```python
data = df.groupby('개봉 연도')['평점'].mean(True)

data
```

```python
plt.plot(data.index, data, marker='o')
plt.xticks([2005,2010,2015,2020])
plt.ylim(7,10)
plt.show()
```

## 3. 평점이 9점 이상인 영화의 비율할수 있는 원그래프를 그리세요

- label : 9점 이상, 9점 미만
- 퍼센트로 표시 소수점 첫째자리까지표시
- 범례 : 그래프 우측 표시

```python
df[df['평점'] >= 9]
df[df['평점'] < 9]

f = df['평점'] >= 9

values = [len(df[f]),len(df[~f])]
labels = ['9점 이상', '9점 미만']
```

```python
plt.pie(values, labels=labels, autopct = '%.1f%%')
plt.legend(loc=(1,0.5))
plt.show()
```

## 4. 산점도 그래프를 그리세요

- 연도별 관객수와 평점을 표시
- x축 : 관객수
- y축 : 평점
- 개봉 연도별로 색깔을 다르게 표시
- 컬러맵 : viridis
- 컬러바도 표시

```python
plt.scatter(df['관객 수'], df['평점'], c= df['개봉 연도'], cmap = 'viridis')
plt.colorbar(label = '개봉 연도')
plt.xlabel('관객 수')
plt.ylabel('평점')
plt.show()
```
