# 산점도 그래프 (scatter chart)

두 변수간의 상관관계

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.pyplot as plt

height = [170,175,180,185,190]
scores = [60,65,70,75,80]

scores.sort(reverse=True)


#산점도 그래프
plt.scatter(height, scores)

#레이블 추가
plt.xlabel('키(cm)')
plt.ylabel('농구점수')

plt.show()
```

## 데이터 프레임 활용

```python
import pandas as pd
df = pd.read_csv('../Pandas/score.csv', encoding = 'utf-8-sig')
df
```

```python
sizes = [100]*8

plt.scatter(df['영어'],df['수학'], s = sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')

plt.title('영어 수학 점수')
plt.show()
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
sizes = df['학년'] * 100

plt.scatter(df['영어'],df['수학'], s = sizes, c= 'r')
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')

plt.title('학년별 영어와 수학 점수')
plt.show()
```

```python
plt.scatter(df['영어'],df['수학'], s = sizes, c = df['학년'], cmap = 'viridis', alpha = 0.7)

#레이블 
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')

#컬러바
cbar = plt.colorbar(ticks=[1,2,3], label = '학년', shrink=0.5, orientation = 'horizontal')
cbar.set_ticklabels(['1학년','2학년','3학년'])

plt.title('학년별 영어와 수학 점수')
plt.show()
```

```python
#컬러맵 확인
from matplotlib import cm

cmaps = plt.colormaps()
for cm in cmaps:
    print(cm)
```

## DataFrame.plot()

```python
df.plot(kind='scatter', x='영어',y='수학', s=sizes, c=df['학년'], colormap = 'viridis')

cmap = plt.cm.get_cmap('viridis')
norm = plt.Normalize(min(df['학년']), max(df['학년']))
sm = plt.cm.ScalarMappable(cmap = cmap, norm=norm)
sm.set_array([])

plt.colorbar(sm, ticks=[1,2,3]).set_ticklabels(['1학년','2학년','3학년'])
plt.show()
```
