# 파일저장

그래프를 이미지 파일로 저장

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[-20,-10,10,20]

plt.plot(x,y)
plt.savefig('graph.png')
```

## 그래프의 크기



```python
plt.figure(figsize=(10,10))
# plt.figure(dpi=200)
plt.plot(x,y)
plt.savefig('graph_size.png')
```

```python
plt.figure(dpi=200)
plt.plot(x,y)
plt.savefig('graph_dpi.png')
```

## 파일 저장 해상도 설정

```python
plt.figure(dpi=50)
plt.plot(x,y)
plt.savefig('graph_savedpi.png', dpi = 200)
```
