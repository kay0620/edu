## COVID-19 백신 종류별 접종 인구

```python
import matplotlib.pyplot as plt

days = [1,2,3,4]
az = [2,4,8,7] #1일부터 4일까지 아스트라제네카 접종인구(단위:만명)
pfizer = [5,1,3,6] #화이자
moderna = [1,2,5,3] #모더나

plt.plot(days, az, marker = 'o', label = '아스트라제네카')
plt.plot(days, pfizer, marker = 'v', label = '화이자')
plt.plot(days, moderna, marker = '*', label = '모더나')

#제목 설정
plt.title('COVID-19 백신 종류별 접종 인구', size = 15)

#눈금 설정 
plt.xticks(days,['%d일'%(i) for i in days]) #['1일','2일','3일','4일'])
plt.yticks(range(0,10,2))

#범위 설정
plt.ylim(0,9)

#레이블 추가
plt.xlabel('날짜(일)')
plt.ylabel('인구수(단위:만명)')

#범례
plt.legend(ncol=3, fontsize = 10, loc = 'upper center')

plt.show()

```

```python

#눈금값 변경
#['1일','2일','3일''4일]
days_labels = []
for i in days:
    days_labels.append('%d일'%(i))
```

```python
days_labels = ['%d일'%(i) for i in days] 
days_labels
```

## 데이터 프레임 활용

```python
import pandas as pd

df = pd.read_csv('../Pandas/score.csv', encoding='utf-8-sig', index_col = '지원번호')
df
```

```python
plt.plot(df['이름'],df['국어'], label = '국어')
plt.plot(df['이름'],df['영어'], label = '영어')
plt.plot(df['이름'],df['수학'], label = '수학')

#제목 설정
plt.title('학생별 성적', size = 15)

#범례
plt.legend(fontsize = 10, loc = 'lower right')

#그리드 설정
plt.grid(ls = ':')

plt.show()
```

```python
plt.plot(df.index,df['국어'])
plt.show()
```

```python
df.plot(kind='line', x='이름', y='키', title = '학생별 키 데이터', legend = False)
plt.ylabel('키(cm)')
plt.xticks(range(len(df)), df['이름'], fontsize = 12)
plt.show()
```

```python
df.plot(kind='line', x='val', y='sin')
plt.show()
```
