## 파일 불러오기

```python
with open('speech_moon.txt', encoding = 'utf-8') as f:
    speech = f.read()
speech
```

## 데이터 전처리

```python
# 불필요한 문자 제거
import re
speech = re.sub('[^가-힣]',' ', speech)
#한글이 아닌 모든 문자를 공백으로 변경
speech
```

```python
from konlpy.tag import Okt

def get_nouns(text):
    nouns = Okt().nouns(text) #명사만 저장
    nouns = [word for word in nouns if len(word) > 1]
    return nouns
```

```python
# 명사 추출
from konlpy.tag import Okt

# nouns = Okt().nouns(speech)
nouns = get_nouns(speech)
nouns
```

## 그래프로 시각화

```python
import pandas as pd
df_word = pd.DataFrame({'word':nouns})
df_word.shape
```

```python
word_freq = df_word.groupby('word').value_counts()
top20 = word_freq.sort_values(ascending=False).head(20)
```

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 #글자크기
# 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.barplot(y = top20.index, x = top20)
plt.show()
```

## 워드 클라우드 생성

```python
dic_word = word_freq.to_dict()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(font_path = 'malgun',
                      background_color = 'white',
                      width = 800, height=500
                     ).generate_from_frequencies(dic_word)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

## 이미지로 워드 클라우드 생성


```python
from PIL import Image 
import numpy as np

img = Image.open('SouthKorea.jpg')
img_array = np.array(img)
```
