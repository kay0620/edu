# 워드클라우드

텍스트 데이터에서 단어들의 빈도를 시각적으로 표현하는 방법 중 하나입니다. <br>
워드 클라우드는 각 단어의 빈도에 따라 크기나 색상을 다르게 설정하여 단어들을 시각화한 이미지를 생성합니다. <br>
이를 통해 텍스트 데이터에서 주요 단어들을 한눈에 파악할 수 있습니다

- 워드클라우드 생성과정
1. 텍스트 데이터 전처리: 텍스트 데이터를 토큰화하고 불용어를 제거하는 등의 전처리 과정을 거칩니다.
1. 단어 빈도 추출: 전처리된 텍스트 데이터에서 각 단어의 빈도를 추출합니다.
1. 워드 클라우드 생성: 추출된 단어 빈도 정보를 기반으로 워드 클라우드를 생성합니다. 단어의 빈도에 따라 단어의 크기나 색상을 다르게 설정하여 시각화합니다.
1. 워드 클라우드 출력 또는 저장: 생성된 워드 클라우드를 출력하거나 이미지 파일로 저장합니다.

```python
# 이미지를 확대 또는 축소할 때 픽셀 간의 보간 방법을 지정하는 속성입니다. 
# 워드 클라우드를 이미지로 표현할 때 interpolation 속성을 사용하여 
# 이미지의 품질과 부드러움을 조절할 수 있습니다.

# 일반적으로 interpolation 속성은 이미지의 픽셀 값을 계산하기 위해 
# 주변 픽셀 값들을 활용합니다. 다양한 보간 방법이 있지만, 
# 주로 사용되는 몇 가지 방법은 다음과 같습니다:

# 'bilinear': 이웃하는 4개의 픽셀 값을 고려하여 새로운 픽셀 값을 계산합니다. 
# 픽셀 간의 색상 변화를 부드럽게 표현합니다.
# 'nearest': 가장 가까운 픽셀의 값을 사용하여 픽셀 값을 계산합니다. 
# 계산이 빠르지만 픽셀 간의 부드러운 변화가 없습니다.
# 'bicubic': 이웃하는 16개의 픽셀 값을 고려하여 새로운 픽셀 값을 계산합니다. 
# 'bilinear'에 비해 더 많은 픽셀 값을 고려하므로 부드러운 이미지를 생성합니다.
# interpolation 속성은 워드 클라우드의 이미지 표현에 적용됩니다. 
# 다양한 속성 값으로 테스트해보면서 적합한 보간 방법을 선택할 수 있습니다.

```

해당 프로그래밍 언어를 이용하는 엔지니어의 수: 과거에 프로그래밍 언어를 배웠던 엔지니어들도 포함되기 때문에 이 수치는 프로그래밍 언어의 과거부터 현재까지의 인기를 간접적으로 포함한다.

Google, Bing, Wikipedia 등에서 프로그래밍 언어가 검색된 횟수: 현재 인기 있는 프로그래밍 언어를 나타낸다.

```python
import requests
url = 'https://www.tiobe.com/tiobe-index/'
re = requests.get(url)
html = re.text
```

```python
from bs4 import BeautifulSoup 
soup = BeautifulSoup(html,'html.parser')
```

```python
elem = soup.select('#top20>tbody>tr')
# print(elem[0])

print(elem[0].select('td'))
print(elem[0].select('td')[4])
print(elem[0].select('td')[5])

print(elem[0].select('td')[4].get_text())
print(elem[0].select('td')[5].get_text())
```

```python
data = {}

for e in elem:
    #print(e.select('td')[4].get_text(),e.select('td')[5].get_text())
    lang = e.select('td')[4].get_text()
    rating = e.select('td')[5].get_text()
#     print(float(rating.replace('%','')))
    rating = float(rating.replace('%',''))
    
    data[lang] = rating

data
```

```python
wordcloud = WordCloud(font_path='malgun',
               #background_color = 'white', 
               colormap='Set2',
               width=800, height=400).generate_from_frequencies(data)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
#데이터프레임 생성

import pandas as pd

prolang_rank = pd.DataFrame(table_rank, columns=['2022','2021','ProgrammingLanguage','Ratings','Change'])
prolang_rank
```

### 빈도수 구하기

```python
from nltk import Text

# 텍스트 데이터 생성
text_data = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']

# nltk.Text() 객체 생성
text = Text(text_data)

# 가장 일반적인 단어들과 빈도수 출력
common_words = text.vocab().most_common()
for word, frequency in common_words:
    print(word, frequency)
```

### 토큰화

- 문장 토큰화:  주어진 텍스트를 문장 단위로 분리하는 작업
- 단어 토큰화: 주어진 텍스트를 단어 단위로 분리하는 작업

```python
from nltk.tokenize import sent_tokenize, word_tokenize

# 문장 토큰화
text = "Hello, how are you? I'm doing well."
sentences = sent_tokenize(text)
print(sentences)

# 단어 토큰화
words = word_tokenize(text)
print(words)
```

### 품사 태깅 (Part-of-Speech Tagging)

문장 내 각 단어에 해당하는 품사(명사,동사,형용사등) 태깅

```python
#품사 태깅(Part-of-Speech Tagging):
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

# 텍스트 데이터
text = "I love coding."

# 품사 태깅
words = word_tokenize(text)
tags = pos_tag(words)
print(tags)
```

### 불용어 처리(Stopword Removal)

불용어
- 일반적으로 의미를 가지지 않거나 분석에 큰 영향을 주지 않는 단어
- a, an, the, in, on, is등과 같이 관사, 전치사,동사등이 해당

주어진 텍스트에서 불용어를 식별하고 제거하여 분석에 중요한 단어들만 추출<br>
분석 결과 정확도 높임

```python


import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# 텍스트 데이터
text = "This is an example sentence demonstrating stop word removal."

# 불용어 제거
stop_words = set(stopwords.words('english'))
words = word_tokenize(text)
filtered_words = [word for word in words if word.lower() not in stop_words]
print(filtered_words)
# print(stop_words)
```

```python
from konlpy.tag import Okt
# 명사 추출 함수
def get_nouns(text):
    okt = Okt()
    nouns = okt.nouns(text)
    nouns = [word for word in nouns if len(word) > 1]  # 한 글자 명사 제외
    return nouns

# 명사 추출
nouns = get_nouns(moon)
```

```python
# 데이터 프레임으로 변환
import pandas as pd
df_word = pd.DataFrame({'word' : nouns})
df_word
```

```python
df_word = df_word.groupby('word').value_counts().sort_values(ascending=False)
```

```python
# 단어 빈도 막대 그래프 그리기

# 단어 빈도 상위 20개 추출
top20 = df_word.head(20)
top20
```

```python
# 형태소 분석(Morphological Analysis):
import nltk
from nltk.stem import WordNetLemmatizer

# 텍스트 데이터
text = "He is running. He ran. He will run."

# WordNetLemmatizer를 사용한 표제어 추출
#단어 원형 복원 
#running", "runs", "ran"과 같은 단어들은 모두 "run"이라는 기본형으로 복원

lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(word) for word in word_tokenize(text)]
print(lemmas)
print(lemmatizer)
```
