# 문자열

```python
import pandas as pd

df = pd.read_csv('score.csv', encoding='utf-8-sig', index_col = '지원번호')
df
```

## 공백제거

- strip() : 앞뒤공백제거
- lstrip() : 앞 공백 제거
- rstrip() : 뒤 공백 제거

```python
#문자열 길이
df['SW특기'].str.len() 
```

```python
#양쪽 공백 제거
df['SW특기'] = df['SW특기'].str.strip()
```

```python
#앞쪽 공백 제거 (C 앞에 공백 제거)
df['SW특기'].str.lstrip().str.len()
```

```python
#뒤쪽 공백 제거 (java,Javascript 뒤쪽 공백 제거 )
df['SW특기'].str.rstrip().str.len()
```

## 검색

- startswith() : 시작하는 글자 찾기
- endswith() : 끝나는 글자 찾기
- contains() : 글자가 포함
- find() : 글자의 위치를 알려주는 함수

```python
#황씨 성을 가진 학생
df['이름'].str.startswith('황')

df[df['이름'].str.startswith('황')]
```

```python
#이름이 '호'로 끝나는 학생
df['이름'].str.endswith('호')

df[df['이름'].str.endswith('호')]
```

```python
#이름에 '태'가 포함된 학생
f=df['이름'].str.contains('태')
df[df['이름'].str.contains('태')]

#이름에 '태'가 들어가지 않는 학생
df[~f]
```

```python
df['SW특기'].str.contains('Java')
df[df['SW특기'].str.contains('Java')] # nan 데이터때문에 오류남
```

```python
df[df['SW특기'].str.contains('Java', na=True)] #nan 데이터에 대해서 True로 설정
```

```python
# 특정 언어를 다룰줄 아는 사람 찾기
df['SW특기'].isin(['Python','java'])
df[df['SW특기'].isin(['Python','java'])]
```

## 대소문자 변경

- lower() : 소문자로 변경
- upper() : 대문자로 변경
- capitalize() : 문장의 첫글자 대문자로
- title() : 단어의 첫글자 대문자로
- swapcase() : 소문자는 대문자로, 대문자는 소문자로 

```python
df['SW특기'].str.lower() #소문자로 변경
df['SW특기'].str.upper() #대문자로 변경
```

```python
#공백 때문에 java 검색이 안됨
df['SW특기'] = df['SW특기'].str.strip()
```

```python
f = df['SW특기'].str.lower()
df[f.isin(['python','java'])]
```

## 어절 나누기

- split()

```python
df.loc['2번','SW특기'] = 'Java Script'
```

```python
df['SW특기'].str.split(' ') #공백으로 나누기
df['SW특기'].str.split(' ', expand= True)
```

## 문자열 바꾸기

- replace()

```python
#특정 문자로 변경
df['SW특기'] = df['SW특기'].str.replace(' ', '/')
df
```

```python
#특정 문자를 삭제
df['SW특기'] = df['SW특기'].str.replace('/','')
```
