# 병합

여러 데이터를 합쳐 하나로 만드는 것

- 가로 병합
- 세로 병합

```python
import pandas as pd

df = pd.read_csv('score.csv', encoding='utf-8-sig', index_col = '지원번호')
df
```

### 가로 병합

- merge()

```python
data = {
    '정보' : [88,55,75,89,99,100,55,64]
}
df.index
df2 = pd.DataFrame(data, index= df.index)
```

```python
#오른쪽 데이터 프레임을 왼쪽 데이터 프레임에 병합

pd.merge(df, df2, how='left', on='지원번호')
```
