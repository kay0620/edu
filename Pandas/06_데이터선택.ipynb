## loc

이름을 이용해서 원하는 row에서 원하는 col 선택

- loc[행]
- loc[행,열]
- loc[조건, 열]

```python

df.loc['5번', '키']
df.loc[['1번','5번'], '키']
df.loc[['1번','5번'], ['이름','키']]
df.loc['1번':'5번', '이름':'키']

df.loc['2번':'5번', '이름':'SW특기']
```

```python
df[df['국어'] >=70]
df.loc[df['국어'] >=70, ['이름','국어']]
```

## iloc

위치를 이용하여 원하는 row에서 원하는 col 선택

- iloc[행]
- iloc[행,열]

```python
df.iloc[3] #4번째 데이터
```

```python
df.iloc[3,1]
df.iloc[3,[0,1]]
df.iloc[[0,3],[0,1]]
df.iloc[[0,3],[0,1,-1]]
```

```python
# 6번 8번 학생의 이름과 국어 점수를 출력하세요

# 6번	변덕규	80
# 8번	윤대협	100

df.iloc[[5,7],[0,3]]
```

```python
df.iloc[:5] #0~4번째 까지 데이터 

df.iloc[:5,0:4] #0~4번째 까지 데이터 , 0~3번째 데이터 (이름:국어)

df.iloc[:5, 3:-1] #0~4번째 학생의 국어~사회점수
df.iloc[:5, 3:]

df.iloc[:, 3]
```

## 조건

```python
df[df['키'] >= 190]
df[df['키'] < 190]
```

```python
f = df['키'] >= 190

df[f] # 키가 190 이상인 학생들
df[~f] #키가 190 미만인 학생들
```

```python
df.loc[df['키'] >= 190, '키'] # 키가 190 이상인 학생들

df.loc[df['키'] >= 190, ['이름','키','학교']] # 키가 190 이상인 학생들
```

```python
#다양한 조건
# & 그리고


(df['키']>=190) & (df['학교'] == '능남고')
df[(df['키']>=190) & (df['학교'] == '능남고')]


df.loc[(df['키']>=190) & (df['학교'] == '능남고'), '키']
df.loc[(df['키']>=190) & (df['학교'] == '능남고'), ['이름','키','학교']]
```

```python
# | 또는
f1 = df['키'] < 170
f2 = df['키'] > 200

f1 | f2
df[f1 | f2]

df.loc[f1|f2, ['이름','학교','키']]
```

```python
# quiz
# 국어 80이상 이면서 영어 80
# 이름, 학교, 국어, 영어
```

```python
df.loc[(df['국어']>=80) & (df['영어'] >=80), ['이름','학교','국어','영어']]
```

```python
f1 = df['국어'] >= 80
f2 = df['영어'] >= 80
f1 & f2
df.loc[f1 & f2, ['이름', '학교', '국어', '영어']]
```
