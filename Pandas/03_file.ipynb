# 파일

excel, csv, txt 형태의 파일로 저장하고 읽어오기



```python
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java ', 'Javascript ', '', '', '  C', 'PYTHON', 'C#']
}
```

```python
import pandas as pd

df = pd.DataFrame(data)
```

## 파일 저장

```python
#txt 파일
df.to_csv('score.txt', sep = '\t')
```

### 인코딩

- utf-8
- utf-8-sig
- cp949

```python
#csv 파일
df.to_csv('score.csv', encoding = 'utf-8-sig', index=False)
```

## 파일 열기

```python
#txt 파일
df = pd.read_csv('score.txt', sep = '\t', index_col = 'Unnamed: 0')
df
```

```python
#csv 파일 
pd.read_csv('score.csv', encoding = 'utf-8-sig', skiprows=3) #지정된 갯수만큼 행을 건너뛰고 내용을 불러옴
```

```python
pd.read_csv('score.csv', encoding = 'utf-8-sig', skiprows=[3,5]) #3,5번째 행을 제외
pd.read_csv('score.csv', encoding = 'utf-8-sig', nrows=4)
pd.read_csv('score.csv', encoding = 'utf-8-sig', skiprows=[2], nrows=4)
```

```python
#엑셀 
pd.read_excel('score.xlsx', index_col='이름')
```
