다음은 대한민국 영화 중에서 관객수가 가장 많은 상위 8개의 데이터입니다.<br>
주어진 코드를 이용해서 퀴즈를 풀어보세요

```python
data = {
    '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
```

1. 데이터 프레임을 생성하세요

```python
import pandas as pd

df = pd.DataFrame(data)
df
```

2. 전체 데이터 중에서 영화 정보만 출력하세요

```python
df['영화']
```

3. '영화', '평점' 정보를 출력하세요

```python
df[['영화', '평점']]
```

4. 2015년 이후에 개봉한 영화 데이터 중에서 '영화','개봉연도' 정보를 출력하세요

```python
df[df['개봉 연도'] >= 2015]
# df.loc[df['개봉 연도'] >= 2015, ['영화','개봉 연도']]

df.loc[df['개봉 연도'] >= 2015, ['영화','개봉 연도']]
```

```python
tmp = df[df['개봉 연도'] >= 2015]
tmp[['영화','개봉 연도']]
```

5. 추천 점수 컬럼을 추가하세요

- 추천 점수 = (관객수*평점)//100

```python
df['추천점수'] = df['관객 수'] * df['평점'] //100
```

```python
df['추천점수'] = df['추천점수'].astype(int)
```

```python
df
```

6. 전체 데이터를 '개봉 연도' 기준으로 정렬해서 출력하세요 (내림차순)

```python
df.sort_values('개봉 연도', ascending = False, inplace = True)

sortdf = df.sort_values('개봉 연도', ascending = False)
sortdf
```

7. 장르를 추가하고 액션이 포함된 영화를 '평점' 기준 내림차순으로 출력하세요

```python
df['장르'] = ['액션/전쟁','코미디/범죄','액션/판타지','드라마/전쟁','액션/공포','드라마/액션','드라마/코미디','드라마/액션']
```

```python
df2 = df.sort_values('평점', ascending = False)
result = df2[df2['장르'].str.contains('액션')]
result
```

```python
df3 = df[df['장르'].str.contains('액션')]
df3.sort_values('평점', ascending=False)
```

```python
df[df['장르'].str.contains('액션')].sort_values('평점', ascending = False)
```

8. 장르를 '/' 기준으로 앞에 내용으로 수정하세요
전쟁/액션 -> 전쟁

```python
tmp = df['장르'].str.split('/', expand=True)
tmp[0]
```

```python
df['장르'] = df['장르'].str.split('/', expand=True)[0]
```

```python
df
```

9. 장르로 그룹화를 하고 장르별로 관객수, 평점, 추천점수의 평균을 출력하세요

```python
df.groupby('장르')[['관객 수','평점','추천점수']].mean(numeric_only = True).round(2)
```
