```python
import pandas as pd

df = pd.read_csv('score.csv', encoding='utf-8-sig', index_col = '지원번호')
df
```

### 학교 이름 변경 

- 북산고 -> 상북고
- 상북고등학교

```python
#데이터 확인
df['학교'].unique()
```

```python
# 북산고 -> 상북고
df['학교'].str.replace('북산고','상북고')
#능남고 -> 남고
df['학교'].replace({'북산고':'상북고','능남고':'남고'}, inplace=True)
df
```

```python
# 상북고 -> 상북고등학교

df['학교'] = df['학교'] + '등학교'
df
```

### SW특기 변경



```python
#대문자로 변경
df['SW특기'] = df['SW특기'].str.upper()
```

```python
df.loc['4번','SW특기'] = 'python'.upper()
```

```python
df.loc['5번',['학교','SW특기']] = ['남고등학교','C']
```

```python
df
```

### 총합과 결과 추가

```python
df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
```

```python
df['총합'] = df.loc[:, '국어':'사회'].sum(axis=1)
```

```python
df
```

```python
#총합이 400점 이상이면 pass, 아니면 fail

df['결과'] = None
df.loc[df['총합'] >= 400, '결과'] = 'Pass'
df.loc[df['총합'] < 400, '결과'] = 'fail'
```

```python
df['결과'] = 'Fail'
df.loc[df['총합'] >= 400, '결과'] = 'Pass'
```

```python
f = df['총합'] >= 400
df.loc[f, '결과'] = 'Pass'
df.loc[~f, '결과'] = 'Fail'
```

```python
import numpy as np

df['결과'] = np.where(df['총합'] >= 400, 'Pass', 'Fail')
```

```python
df
```

```python
df.drop(labels='결과', axis=1)
#axis : 방향 (1:열, 0:행)
```

```python
df.drop(columns=['결과','사회'])
df.drop(index = ['4번','6번'])
```
